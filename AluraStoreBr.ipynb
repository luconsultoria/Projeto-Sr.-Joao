### Importação dos dados



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

# Leitura dos dados
df_loja_1 = pd.read_csv(url)
df_loja_2 = pd.read_csv(url2)
df_loja_3 = pd.read_csv(url3)
df_loja_4 = pd.read_csv(url4)

# Preparação
dfs_lojas = [df_loja_1, df_loja_2, df_loja_3, df_loja_4]
nomes_lojas = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]
dfs_lojas = [
    df_loja_1[["Categoria do Produto", "Avaliação da compra"]],
    df_loja_2[["Categoria do Produto", "Avaliação da compra"]],
    df_loja_3[["Categoria do Produto", "Avaliação da compra"]],
    df_loja_4[["Categoria do Produto", "Avaliação da compra"]],
]

# Criando a coluna Faturamento antes de usá-la
df_loja_1["Faturamento"] = df_loja_1["Preço"]
df_loja_2["Faturamento"] = df_loja_2["Preço"]
df_loja_3["Faturamento"] = df_loja_3["Preço"]
df_loja_4["Faturamento"] = df_loja_4["Preço"]

df_faturamento_total_da_loja_1 = pd.DataFrame({"Faturamento Total Loja 1 (R$)": [df_loja_1["Faturamento"].sum()]})
df_faturamento_total_da_loja_2 = pd.DataFrame({"Faturamento Total Loja 2 (R$)": [df_loja_2["Faturamento"].sum()]})
df_faturamento_total_da_loja_3 = pd.DataFrame({"Faturamento Total Loja 3 (R$)": [df_loja_3["Faturamento"].sum()]})
df_faturamento_total_da_loja_4 = pd.DataFrame({"Faturamento Total Loja 4 (R$)": [df_loja_4["Faturamento"].sum()]})

loja.head()
```

#1. Análise do faturamento


```python

faturamento = [
    df_loja_1["Faturamento"].sum(),
    df_loja_2["Faturamento"].sum(),
    df_loja_3["Faturamento"].sum(),
    df_loja_4["Faturamento"].sum()
]

df_faturamento_total_da_loja_1 = pd.DataFrame({"Faturamento Total Loja 1 (R$)": [df_loja_1["Faturamento"].sum()]})
print(df_faturamento_total_da_loja_1, "\n")

df_faturamento_total_da_loja_2 = pd.DataFrame({"Faturamento Total Loja 2 (R$)": [df_loja_2["Faturamento"].sum()]})
print(df_faturamento_total_da_loja_2, "\n")

df_faturamento_total_da_loja_3 = pd.DataFrame({"Faturamento Total Loja 3 (R$)": [df_loja_3["Faturamento"].sum()]})
print(df_faturamento_total_da_loja_3, "\n")

df_faturamento_total_da_loja_4 = pd.DataFrame({"Faturamento Total Loja 4 (R$)": [df_loja_4["Faturamento"].sum()]})
print(df_faturamento_total_da_loja_4, "\n")

plt.figure(figsize=(8, 5))
plt.bar(nomes_lojas, faturamento, color=['skyblue', 'lightcoral', 'lightgreen', 'yellow'])
plt.title("Faturamento Total por Loja")
plt.ylabel("Faturamento (R$)",)
plt.xlabel("Lojas",)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# 2. Vendas por Categoria


```python
# Categorias Mais Vendidas
df_categorias_mais_vendidas_da_loja_1 = df_loja_1["Categoria do Produto"].value_counts().reset_index()
df_categorias_mais_vendidas_da_loja_1.columns = ["Categoria", "Unidades Vendidas"]
print(df_categorias_mais_vendidas_da_loja_1.head(), "\n")

df_categorias_mais_vendidas_da_loja_2 = df_loja_2["Categoria do Produto"].value_counts().reset_index()
df_categorias_mais_vendidas_da_loja_2.columns = ["Categoria", "Unidades Vendidas"]
print(df_categorias_mais_vendidas_da_loja_2.head(), "\n")

df_categorias_mais_vendidas_da_loja_3 = df_loja_3["Categoria do Produto"].value_counts().reset_index()
df_categorias_mais_vendidas_da_loja_3.columns = ["Categoria", "Unidades Vendidas"]
print(df_categorias_mais_vendidas_da_loja_3.head(), "\n")

df_categorias_mais_vendidas_da_loja_4 = df_loja_4["Categoria do Produto"].value_counts().reset_index()
df_categorias_mais_vendidas_da_loja_4.columns = ["Categoria", "Unidades Vendidas"]
print(df_categorias_mais_vendidas_da_loja_4.head(), "\n")

for i, df in enumerate(dfs_lojas):
    categorias = df["Categoria do Produto"].value_counts().head(10)
    plt.figure(figsize=(6, 6))
    plt.pie(categorias, labels=categorias.index, autopct='%1.1f%%', startangle=140)
    plt.title(f"Categorias Mais Vendidas - {nomes_lojas[i]}")
    plt.tight_layout()
    plt.show()

# Categorias Mais Rentáveis
df_categorias_mais_rentaveis_da_loja_1 = (
    df_loja_1.groupby("Categoria do Produto")["Faturamento"]
    .sum()
    .reset_index()
    .sort_values(by="Faturamento", ascending=False)
    .rename(columns={"Categoria do Produto": "Categoria"})
)
print(df_categorias_mais_rentaveis_da_loja_1.head(), "\n")

df_categorias_mais_rentaveis_da_loja_2 = (
    df_loja_2.groupby("Categoria do Produto")["Faturamento"]
    .sum()
    .reset_index()
    .sort_values(by="Faturamento", ascending=False)
    .rename(columns={"Categoria do Produto": "Categoria"})
)
print(df_categorias_mais_rentaveis_da_loja_2.head(), "\n")

df_categorias_mais_rentaveis_da_loja_3 = (
    df_loja_3.groupby("Categoria do Produto")["Faturamento"]
    .sum()
    .reset_index()
    .sort_values(by="Faturamento", ascending=False)
    .rename(columns={"Categoria do Produto": "Categoria"})
)
print(df_categorias_mais_rentaveis_da_loja_3.head(), "\n")

df_categorias_mais_rentaveis_da_loja_4 = (
    df_loja_4.groupby("Categoria do Produto")["Faturamento"]
    .sum()
    .reset_index()
    .sort_values(by="Faturamento", ascending=False)
    .rename(columns={"Categoria do Produto": "Categoria"})
)
print(df_categorias_mais_rentaveis_da_loja_4.head(), "\n")

for i, df in enumerate([df_loja_1, df_loja_2, df_loja_3, df_loja_4]):
    rentabilidade = df.groupby("Categoria do Produto")["Faturamento"].sum().sort_values(ascending=False).head(10)
    plt.figure(figsize=(8, 5))
    rentabilidade.plot(kind='bar', color=['mistyrose', 'palegreen', 'orchid', 'teal','darksalmon','steelblue','lightgrey','turquoise','khaki', 'oldlace'])
    plt.title(f"Categorias Mais Rentáveis - {nomes_lojas[i]}")
    plt.ylabel("Faturamento (R$)")
    plt.xlabel("Categoria")
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```

# 3. Média de Avaliação das Lojas

```python
# Calcular a média das avaliações da compra por loja
df_media_avaliacoes_loja = {}

for nome_loja, df_loja in zip(nomes_lojas, dfs_lojas):
    media_loja = df_loja["Avaliação da compra"].mean()
    df_media_avaliacoes_loja[f"Média Avaliações {nome_loja}"] = [media_loja]

df_media_avaliacoes_loja = pd.DataFrame(df_media_avaliacoes_loja)
print("Média das Avaliações da Compra por Loja:\n", df_media_avaliacoes_loja, "\n")

avaliacoes = [df["Avaliação da compra"].mean() for df in dfs_lojas]

plt.figure(figsize=(10, 6))
plt.plot(df_media_avaliacoes_loja.columns, df_media_avaliacoes_loja.iloc[0], marker='o', linestyle='-')
plt.title('Média das Avaliações da Compra por Loja')
plt.xlabel('Loja')
plt.ylabel('Média da Avaliação')
plt.grid(True)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Calcular a média da avaliação por categoria do produto por loja
df_media_avaliacao_categoria_do_produto = {}

for nome_loja, df_loja in zip(nomes_lojas, dfs_lojas):
    media_categoria_loja = df_loja.groupby("Categoria do Produto")["Avaliação da compra"].mean().reset_index()
    df_media_avaliacao_categoria_do_produto[nome_loja] = media_categoria_loja
    print(f"Média da Avaliação por Categoria do Produto - {nome_loja}:\n", media_categoria_loja.head(), "\n")

for i, df in enumerate(dfs_lojas):
    media_cat = df.groupby("Categoria do Produto")["Avaliação da compra"].mean().reset_index()
    plt.figure(figsize=(8, 5))
    plt.scatter(media_cat["Categoria do Produto"], media_cat["Avaliação da compra"], color='cyan')
    plt.title(f"Média de Avaliação por Categoria - {nomes_lojas[i]}")
    plt.ylabel("Nota Média")
    plt.xlabel("Categoria")
    plt.xticks(rotation=45)
    plt.ylim(0, 5)
    plt.grid(True, linestyle='--', alpha=0.6)
```
